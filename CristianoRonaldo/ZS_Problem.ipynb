```python
for i in float_cols:
    print(i, "\n", df[i].value_counts(),"\n")
```

```python
to_pop=["is_goal","location_x","location_y"]
for i in to_pop:
    float_cols.remove(i)
df[float_cols]=df[float_cols].fillna(-1)
```

```python
df["location_x"]=df["location_x"].fillna(-1000)
df["location_y"]=df["location_y"].fillna(-1000)
```

```python
correlations = df.corr()['is_goal'].drop('is_goal')
print(correlations)
```

```python
a_corr=abs(correlations)
features=a_corr[a_corr>0].index.values.tolist()
len(features)
```

```python
df["is_goal"]=df["is_goal"].fillna(5)
df=df[df["is_goal"]!=5]
```

```python
import numpy as np
x=df[features]
y=np.asarray(df['is_goal'])
```

```python
from sklearn import preprocessing
x=preprocessing.StandardScaler().fit_transform(x) #standardizing
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=1)
print ('Train set:', x_train.shape,  y_train.shape)
print ('Test set:', x_test.shape,  y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression( solver='newton-cg',multi_class='multinomial')
model=LR.fit(x_train,y_train)
```

```python
y_pred_train=model.predict(x_train)
y_pred_test=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_train,y_pred_train)
```

```python
accuracy_score(y_test,y_pred_test)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y_train,y_pred_train)
```

```python
mean_absolute_error(y_test,y_pred_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=5,random_state=1)
rf.fit(x_train, y_train)
rf_predict=rf.predict(x_test)
rf_acc_score = accuracy_score(y_test, rf_predict)
print(rf_acc_score*100)
```

```python
accuracy_score(y_train, rf.predict(x_train))
```

```python
mean_absolute_error(y_train,rf.predict(x_train))
```
