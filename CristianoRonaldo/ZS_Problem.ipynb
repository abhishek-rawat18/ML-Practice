```python
shot_basics=df["shot_basics"].unique()
shot_basics=[x for x in shot_basics if str(x)!="nan"]
range_of_shot=df["range_of_shot"].unique()
range_of_shot=[x for x in range_of_shot if str(x)!="nan"]

shot_basics_dist_dict=dict() #just to store the min and max values of distance_of_shot corresponding to shot_basics
range_of_shot_dict=dict()
for i in shot_basics:
    sub=df[df["shot_basics"]==i]
    mn=min(sub["distance_of_shot"])
    mx=max(sub["distance_of_shot"])
    shot_basics_dist_dict[i]=[mn,mx]
for i in range_of_shot:
    sub=df[df["range_of_shot"]==i]
    mn=min(sub["distance_of_shot"])
    mx=max(sub["distance_of_shot"])
    range_of_shot_dict[i]=[mn,mx]
```

```python
# now filling missing values for distance of shot by distance of shot.1 if it lies in the limits 
#otherwise the mean of distance of shot

count1=0 #just a counter to note how many values got substituted by mean
count2=0 #how many got substituted by directly ds1 (i.e. no shot basic available corr to that)

for i in range(len(df)):
    ds=df["distance_of_shot"].iloc[i]
    
    if(str(ds)=="nan"):
        
        ds1=df["distance_of_shot.1"].iloc[i]
        sb=df["shot_basics"].iloc[i]
        rs=df["range_of_shot"].iloc[i]
        
        if (str(sb)!="nan"): #if shot basics is not nan then only check if within range, else just substitute by ds.1 value
            
            mn=shot_basics_dist_dict[sb][0]
            mx=shot_basics_dist_dict[sb][1]
            
            if((str(ds1)!="nan") and mn<=ds1 and ds1<=mx):
                
                df["distance_of_shot"].iloc[i]=ds1
                
            else: #if ds1 not in range or nan, substitute by mean
                
                df["distance_of_shot"].iloc[i]=(mn+mx)/2
                count1+=1
        
        else:
            
            if(str(ds1)!="nan"):
                df["distance_of_shot"].iloc[i]=ds1
                count2+=1
            
            elif(str(rs)!="nan"):
                mn=range_of_shot_dict[rs][0]
                mx=range_of_shot_dict[rs][1]
                df["distance_of_shot"].iloc[i]=(mx-mn)/2
```
